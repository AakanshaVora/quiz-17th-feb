## Data science
</div>

With Colab you can harness the full power of popular Python libraries to analyze and visualize data. The code cell below uses **numpy** to generate some random data, and uses **matplotlib** to visualize it. To edit the code, just click the cell and start editing.

You can import your own data into Colab notebooks from your Google Drive account, including from spreadsheets, as well as from Github and many other sources. To learn more about importing data, and how Colab can be used for data science, see the links below under [Working with Data](#working-with-data).

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "/content/listings_detailed.csv"
df = pd.read_csv(file_path)

# Identify potential host/user ID column
host_col = [col for col in df.columns if 'host' in col.lower() and 'id' in col.lower()]
if host_col:
    host_col = host_col[0]  # Take the first matching column
    print(f"Using column: {host_col} for host analysis.")
else:
    raise ValueError("No host/user ID column found in the dataset.")

# Count the number of listings per host
host_listing_counts = df[host_col].value_counts()

# Summary statistics
single_property_owners = (host_listing_counts == 1).sum()
multiple_property_owners = (host_listing_counts > 1).sum()
max_properties_owned = host_listing_counts.max()
mean_properties_owned = host_listing_counts.mean()
median_properties_owned = host_listing_counts.median()

# Calculate the percentage of single vs multiple property owners
total_hosts = len(host_listing_counts)
percentage_single_owners = (single_property_owners / total_hosts) * 100
percentage_multiple_owners = (multiple_property_owners / total_hosts) * 100

# --- Automated Interpretation ---
print("\n--- Interpretation ---")
print(f"Total unique hosts: {total_hosts}")
print(f"Hosts with a single listing: {single_property_owners} ({percentage_single_owners:.2f}%)")
print(f"Hosts with multiple listings: {multiple_property_owners} ({percentage_multiple_owners:.2f}%)")
print(f"On average, a host owns {mean_properties_owned:.2f} properties.")
print(f"The median number of listings per host is {median_properties_owned}.")
print(f"The maximum number of properties owned by a single host is {max_properties_owned}.")

if max_properties_owned > 10:
    print("⚠️ Some hosts own more than 10 properties, indicating possible commercial operations.")
elif max_properties_owned > 5:
    print("⚠️ A few hosts own multiple properties, suggesting they might be operating as businesses.")
else:
    print("✅ Most hosts appear to be private individuals renting a single property.")

if percentage_multiple_owners > 30:
    print("⚠️ More than 30% of hosts own multiple properties. This suggests a significant presence of professional landlords or speculators.")
else:
    print("✅ The majority of hosts own only one property, indicating a primarily private-user market.")

# --- Visualizations ---

# 1. Pie Chart: Single vs. Multiple Property Owners
plt.figure(figsize=(8, 8))
labels = ['Single Listing Owners', 'Multiple Listing Owners']
sizes = [single_property_owners, multiple_property_owners]
colors = ['#66b3ff', '#ff9999']
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140, wedgeprops={'edgecolor': 'black'})
plt.title("Proportion of Single vs. Multiple Property Owners")
plt.show()

# 2. Box Plot: Detecting Outliers in Host Listings
plt.figure(figsize=(10, 6))
sns.boxplot(x=host_listing_counts, showfliers=True, color="lightblue")
plt.xlabel("Number of Listings per Host")
plt.title("Box Plot of Listings per Host (Outlier Detection)")
plt.show()

# 3. Top 10 Hosts with Most Listings (Bar Chart)
plt.figure(figsize=(12, 6))
top_hosts = host_listing_counts.head(10)
sns.barplot(x=top_hosts.index.astype(str), y=top_hosts.values, palette="coolwarm")
plt.xticks(rotation=45)
plt.xlabel("Host ID")
plt.ylabel("Number of Listings")
plt.title("Top 10 Hosts with the Most Listings")
plt.show()

# Print the Top 10 Hosts with the Most Listings
print("\nTop 10 Hosts with the Most Listings:")
print(top_hosts)
```
